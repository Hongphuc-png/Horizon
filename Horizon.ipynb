```python
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.tree import DecisionTreeRegressor
from sklearn.model_selection import train_test_split
import pandas as pd

# Load the data and define features (from previous successful cells)
data = pd.read_csv("/content/Sales-Performance.1.csv")
features = ("Channel", "Preferred Com Channel", "Category Name", "Location", "Product Name", "Age", "Gender")
x = data[list(features)]
y = data["Total Price"] # Assuming "Total_Price" is the target variable

# Perform the train-test split (from cell -6VGp1dqCXbw)
x_train, x_valid, y_train, y_valid = train_test_split(x, y, train_size=0.8, test_size=0.2, random_state=0)

# Identify categorical features
# Based on the columns in sample_data/california_housing_test.csv, there are no object type columns for one-hot encoding.
# Keeping the code for demonstration purposes, but it will likely result in an empty categorical_features list.
categorical_features = x_train.select_dtypes(include=['object']).columns

# Create a ColumnTransformer to apply one-hot encoding to categorical features
# If categorical_features is empty, this preprocessor will effectively do nothing to the numerical columns.
preprocessor = ColumnTransformer(
    transformers=[
        ('onehot', OneHotEncoder(handle_unknown='ignore'), categorical_features)
    ],
    remainder='passthrough' # Keep other columns (if any)
)

# Create a pipeline that first preprocesses the data and then trains the model
model_pipeline = Pipeline(steps=[('preprocessor', preprocessor),
                                 ('regressor', DecisionTreeRegressor(random_state=1))])

# Fit the pipeline on the training data
model_pipeline.fit(x_train, y_train)
```

#Make Predictions

```python
# Make predictions on the validation data using the trained pipeline
y_predict = model_pipeline.predict(x_valid)
display(y_predict[:5]) # Display the first 5 predictions
```

#Evaluate Model

```python
from sklearn.metrics import mean_absolute_error

# Calculate the Mean Absolute Error (MAE) on the validation data
mae = mean_absolute_error(y_valid, y_predict)
print(f"Mean Absolute Error (MAE): {mae}")
```

```python
import pandas as pd
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.tree import DecisionTreeRegressor
from sklearn.model_selection import train_test_split

# Load the data and define features (from previous successful cells)
data = pd.read_csv("/content/Sales-Performance.1.csv")
features = ("Channel", "Preferred Com Channel", "Category Name", "Location", "Product Name", "Age", "Gender")
x = data[list(features)]
y = data["Total Price"] # Assuming "Total Price" is the target variable

# Perform the train-test split (from cell -6VGp1dqCXbw)
x_train, x_valid, y_train, y_valid = train_test_split(x, y, train_size=0.8, test_size=0.2, random_state=0)

# Identify categorical features
# Based on the columns in /Sales-Performance.1.csv, there are no object type columns for one-hot encoding.
# Keeping the code for demonstration purposes, but it will likely result in an empty categorical_features list.
categorical_features = x_train.select_dtypes(include=['object']).columns

# Create a ColumnTransformer to apply one-hot encoding to categorical features
# If categorical_features is empty, this preprocessor will effectively do nothing to the numerical columns.
preprocessor = ColumnTransformer(
    transformers=[
        ('onehot', OneHotEncoder(handle_unknown='ignore'), categorical_features)
    ],
    remainder='passthrough' # Keep other columns (if any)
)

# Create a pipeline that first preprocesses the data and then trains the model
model_pipeline = Pipeline(steps=[('preprocessor', preprocessor),
                                 ('regressor', DecisionTreeRegressor(random_state=1))])

# Fit the pipeline on the training data
model_pipeline.fit(x_train, y_train)

# Define features for the new input using column names from the California housing dataset
features_for_new_input = ["Channel", "Preferred Com Channel", "Category Name", "Location", "Product Name", "Age", "Gender"]

# Create a new input DataFrame with realistic values for the California housing features
new_input = pd.DataFrame([['SUPERMARKET', 'EMAIL', 'DRINKING_YOGURT', 'HANOI', 'Fruity 180', '27', 'MALE']], columns=features_for_new_input)

# Make prediction using the correct pipeline name
model_pipeline.predict(new_input)
```
