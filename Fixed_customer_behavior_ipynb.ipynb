```python
import pandas as pd

df = pd.read_csv('/content/Fixed data demo.csv')
by_Channel_df = df.loc[
    df['Response'] == 'Yes'
].groupby([
    'Channel'
]).count()['Customer Id']/df.groupby('Channel').count()['Customer Id']

by_Channel_df
```

```python
import pandas as pd

df = pd.read_csv('/content/Fixed data demo.csv')
by_Channel_df = df.loc[
    df['Response'] == 'Yes'
].groupby([
    'Channel'
]).count()['Gender']/df.groupby('Channel').count()['Gender']

by_Channel_df
```

```python
import pandas as pd

df = pd.read_csv('/content/Fixed data demo.csv')
by_Channel_df = df.loc[
    (df['Response'] == 'Yes') & (df['Gender'] == 'FEMALE')
].groupby([
    'Channel'
]).count()['Gender']/df.loc[df['Gender'] == 'FEMALE'].groupby('Channel').count()['Gender']

by_Channel_df
```

```python
import pandas as pd

df = pd.read_csv('/content/Fixed data demo.csv')
by_Channel_df = df.loc[
    (df['Response'] == 'Yes') & (df['Gender'] == 'MALE')
].groupby([
    'Channel'
]).count()['Gender']/df.loc[df['Gender'] == 'MALE'].groupby('Channel').count()['Gender']

by_Channel_df
```

```python
import matplotlib.pyplot as plt

ax = (by_Channel_df*100.0).plot(
    kind='bar',
    figsize=(7, 7),
    color='red',
    grid=True
)

ax.set_ylabel('Engagement Rate (%)')

plt.show()
```

**Channel & Category Name**

```python
import matplotlib.pyplot as plt
```

```python
by_Channel_df = df.loc[
    (df['Response'] == 'Yes') & (df['Gender'] == 'FEMALE')
].groupby([
    'Channel', 'Category Name'
]).count()['Gender']/df.loc[df['Gender'] == 'FEMALE'].groupby('Channel').count()['Gender']

by_Channel_df
```

```python
by_Channel_df = df.loc[
    df['Response'] == 'Yes'
].groupby([
    'Channel', 'Category Name'
]).count()['Customer Id']/df.groupby('Channel').count()['Customer Id']

by_Channel_df
```

```python

ax = (by_Channel_df*100.0).plot(
    kind='bar',
    figsize=(10, 7),
    grid=True
)

ax.set_ylabel('Engagement Rate (%)')

plt.show()
```
