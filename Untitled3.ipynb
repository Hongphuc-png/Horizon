**1. Load data**

```python
import pandas as pd
```

```python
df_data = pd.read_csv('/content/Demo.2.csv')
df_data.head()
```

```python
df_data.info()
```

**2. Analytics**

```python
df_data.loc[df_data.Churn=='No','Churn'] = 0
df_data.loc[df_data.Churn=='Yes','Churn'] = 1
```

```python
df_data['Churn'] = df_data['Response'].apply(lambda x: 1 if x == 'Yes' else 0)
df_data.head()
```

```python
df_plot = df_data.groupby('Gender').Churn.mean().reset_index()
plot_data = [
    go.Bar(
        x=df_plot['Gender'],
        y=df_plot['Churn'],
        width = [0.5, 0.5],
        marker=dict(
        color=['green', 'blue'])
    )
]
plot_layout = go.Layout(
        xaxis={"type": "category"},
        yaxis={"title": "Churn Rate"},
        title='Gender',
        plot_bgcolor  = 'rgb(243,243,243)',
        paper_bgcolor  = 'rgb(243,243,243)',
    )
fig = go.Figure(data=plot_data, layout=plot_layout)
pyoff.iplot(fig)
```
