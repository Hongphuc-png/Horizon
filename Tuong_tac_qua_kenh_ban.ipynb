#*2. Analytics on Engaged Customers*


**Engagement Rates by Sales Channel**

```python
df = pd.read_csv('/content/Demo.2.csv')
```

```python
import pandas as pd
```

```python
df = pd.read_csv('/content/Demo.2.csv')
```

```python
by_Channel_df = df.loc[
    df['Response'] == 'Yes'
].groupby([
    'Channel'
]).count()['Customer Id']/df.groupby('Channel').count()['Customer Id']

by_Channel_df
```

```python
ax = (by_Channel_df*100.0).plot(
    kind='bar',
    figsize=(7, 7),
    color='red',
    grid=True
)

ax.set_ylabel('Engagement Rate (%)')

plt.show()
```

**Channel & Category Name**

```python
by_Channel_df = df.loc[
    df['Response'] == 'Yes'
].groupby([
    'Channel', 'Category Name'
]).count()['Customer Id']/df.groupby('Channel').count()['Customer Id']

by_Channel_df
```

```python
ax = (by_Channel_df*100.0).plot(
    kind='bar',
    figsize=(10, 7),
    grid=True
)

ax.set_ylabel('Engagement Rate (%)')

plt.show()
```
